```python
import pandas as pd
print("Success!")
```

```python
df = pd.read_csv('sales_data.csv')

 
print("Data loaded successfully:")
display(df.head())

```

```python
roduct_sales = df.groupby('Product')['Sales'].sum().reset_index()


product_sales = product_sales.sort_values(by='Sales', ascending=False)


display(product_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))


sns.barplot(x='Product', y='Sales', data=product_sales, hue='Product', palette='viridis')


plt.title('Total Sales by Product', fontsize=16)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)


plt.show()-
```
